```python
import pandas as pd
```

```python
# load dataset from data/universal_top_spotify_songs.csv (https://www.kaggle.com/datasets/asaniczka/top-spotify-songs-in-73-countries-daily-updated)
df = pd.read_csv('data/universal_top_spotify_songs.csv')
df.head()
```

```python
# from col snapshot_date, get rows with snapshot_date equal to 2025-06-11 only
df = df[df['snapshot_date'] == '2025-06-11']
df.shape
```

```python
# save the 2025-06-11 dataset to data/spotify_2025_06_11.csv in top50_dataset_2025.csv
df.to_csv('data/spotify_2025_06_11.csv', index=False)
```
